```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
```

## Player Count

* Display the total number of players


```python
purchase_data.columns
```

```python
purchase_data["SN"].unique()
```

```python
total_player = purchase_data["SN"].nunique()

total_player_df = pd.DataFrame({"Total Players": [total_player]})

total_player_df.head()
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
unique_item = purchase_data["Item ID"].nunique()

unique_item
```

```python
average = round(purchase_data["Price"].mean(),2)

average
```

```python
number_purchase = purchase_data["Item ID"].count()

number_purchase
```

```python
total_revenue = purchase_data["Price"].sum()

total_revenue
```

```python
purchase_df = pd.DataFrame({"Number of Unique Items": [unique_item], 
                            "Average Price": [average], 
                            "Number of Purchases": [number_purchase],
                           "Total Revenue": [total_revenue]})

purchase_df["Average Price"] = purchase_df["Average Price"].map('${:,.2f}'.format)
purchase_df["Total Revenue"] = purchase_df["Total Revenue"].map('${:,.2f}'.format)
purchase_df.head()
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
gender_player_df = purchase_data.groupby(["Gender"])

gender_player_df.head()
```

```python
gen_count = gender_player_df["SN"].nunique()

gen_count.head()
```

```python
gen_percent = gen_count/total_player * 100

gen_percent
```

```python
gender_chart = pd.DataFrame({"Total Count": gen_count, "Percentage of Players": gen_percent})

gender_chart = gender_chart.sort_values("Total Count", ascending = False)

gender_chart["Percentage of Players"] = gender_chart["Percentage of Players"].map("{:.2f}%".format)

gender_chart.head()
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting

```python
number_gen_purchase = gender_player_df["Item ID"].count()

number_gen_purchase
```

```python
purchase_analysis_chart = pd.DataFrame({"Purchase Count": number_gen_purchase, 
                                        "Average Purchase Price": aver_gen_price,
                                        "Total Purchase Value": total_gen_purchase,
                                        "Avg Total Purchase per Person": avg_total_person})

purchase_analysis_chart["Average Purchase Price"] = purchase_analysis_chart["Average Purchase Price"].map('${:,.2f}'.format)
purchase_analysis_chart["Total Purchase Value"] = purchase_analysis_chart["Total Purchase Value"].map('${:,.2f}'.format)
purchase_analysis_chart["Avg Total Purchase per Person"] = purchase_analysis_chart["Avg Total Purchase per Person"].map('${:,.2f}'.format)

purchase_analysis_chart.head()
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
bins = [0, 9, 14, 19, 24, 29, 34, 39, 100]

groups_name = ["<10", "10-14", "15-19","20-24","25-29","30-34", "35-39", "40+"]
```

```python
age_player_df = purchase_data.loc[:,["Gender", "SN", "Age"]].drop_duplicates()

age_player_df["Age Range"] = pd.cut(age_player_df["Age"], bins, labels=groups_name)

age_player_df.head()
```

```python
age_count = age_player_df["Age Range"].value_counts()

age_count
```

```python
age_percent = round(age_count/total_player * 100, 2)

age_percent
```

```python
age_chart = pd.DataFrame({"Total Count": age_count, "Percentage of Players": age_percent})

age_chart["Percentage of Players"] = age_chart["Percentage of Players"].map("{:.2f}%".format)

age_chart = age_chart.sort_index()

age_chart.head(10)
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
spender_player_df = purchase_data.groupby("SN")

spender_player_df.head()
```

```python
spender_count = spender_player_df["Price"].count()

spender_count.head()
```

```python
aver_spend_purchase = round(spender_player_df["Price"].mean(), 2)

aver_spend_purchase.head()
```

```python
total_spend_purchase = round(spender_player_df["Price"].sum(), 2)

total_spend_purchase.head()
```

```python
spender_chart = pd.DataFrame({"Purchase Count": spender_count, 
                              "Average Purchase Price": aver_spend_purchase, 
                              "Total Purchase Value": total_spend_purchase})

spender_chart = spender_chart.sort_values(["Total Purchase Value"], ascending = False)

spender_chart["Average Purchase Price"] = spender_chart["Average Purchase Price"].map("${:,.2f}".format)
spender_chart["Total Purchase Value"] = spender_chart["Total Purchase Value"].map("${:,.2f}".format)

spender_chart.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
popular_items_df = purchase_data.groupby(["Item ID", "Item Name"])

popular_items_df.head()
```

```python
item_count = popular_items_df["Price"].count()

item_count.head()
```

```python
popular_item = round(popular_items_df["Price"].mean(), 2)

popular_item.head()
```

```python
item_price_purchase = round(popular_items_df["Price"].sum(),2)

item_price_purchase.head()
```

```python
pop_item_chart = pd.DataFrame({"Purchase Count": item_count, 
                               "Item Price": popular_item, 
                               "Total Purchase Value": item_price_purchase})

pop_item_chart_sort = pop_item_chart.sort_values(["Purchase Count"],ascending = False)

pop_item_chart_sort["Item Price"] = pop_item_chart_sort["Item Price"].map("${:,.2f}".format)
pop_item_chart_sort["Total Purchase Value"] = pop_item_chart_sort["Total Purchase Value"].map("${:,.2f}".format)

pop_item_chart_sort.head()
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
pop_item_chart = pop_item_chart.sort_values(["Total Purchase Value"], ascending = False)

pop_item_chart["Item Price"] = pop_item_chart["Item Price"].map("${:,.2f}".format)
pop_item_chart["Total Purchase Value"] = pop_item_chart["Total Purchase Value"].map("${:,.2f}".format)

pop_item_chart.head()
```
